The loc() function is label based data selecting method which means that we have to pass the name of the row or column which we want to select. This method includes the last element of the range passed in it, unlike iloc(). loc() can accept the boolean data unlike iloc(). Many operations can be performed using the loc() method like

The iloc() function is an indexed-based selecting method which means that we have to pass an integer index in the method to select a specific row/column. This method does not include the last element of the range passed in it unlike loc(). iloc() does not accept the boolean data unlike loc(). Operations performed using iloc() are:

## **Filtering - Conditionals**

We can filter our dataframe as below. We can do the same with loc as we seen before. That's the one reason why Pandas can a bit confusing for the people.

````python
df['Genre']=='Drama'
````

````python
filter = df['Genre']=='Drama'
df.loc[filter]
````

````python
df[df['Genre']=='Drama']
````

WE can not use "and" and "or" operator. So we use symbols instead of them. These symbols are "&" for "and", "|" for "or".

````python
df_temp =df[(df['Language'] == 'German') | (df['Language'] == 'Dutch')]
df_temp
````

````python
df_temp[(df_temp['Genre'] == 'Drama') & (df_temp['month']=='November') & (df_temp['IMDB Score'] >= 5.5)]
````

It is so similiar with filtering as we seen before. But Groupby is different because instead of taking on genre, groupby take all genres and split them.

Instead of getting just one group with filtering, we can get all groups with groupby.

## **Fill NA**

A common way to replace empty cells, is to calculate the mean, median or mode value of the column.

````python
df_temp = df.copy()
data = {'Title': 'Notre Dame de Paris', 'Genre': 'Musical', 'Premiere': "September 18, 1998", "IMDB Score": 8.9, "Language": "French"}
df_temp = df_temp.append(data, ignore_index = True)
df_temp.tail()
````

````python
df_temp['Runtime'].describe()
````

````python
df_temp["Runtime"].fillna(df_temp["Runtime"].mean(), inplace=True)
df_temp.tail()
````

````python
df_temp['year'].fillna(1998, inplace=True)
df_temp['month'].fillna('September', inplace=True)
df_temp.tail()
````
